### OpenCV

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
facial_detector = cv2.CascadeClassifier('/content/drive/MyDrive/Deep Learning/CV2/haarcascade_frontalface_default.xml')
```

```python
image = cv2.imread('/content/drive/MyDrive/Deep Learning/work-space.png')

# Convert to gray
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#Change the scale
new_scale = tuple(round(i / 1.5) for i in reversed(image_gray.shape))
image_gray = cv2.resize(image_gray, new_scale)
image_color = cv2.resize(image, new_scale)

detection = facial_detector.detectMultiScale(image_gray, scaleFactor=1.03, minNeighbors=3, maxSize=(32,32))

for x, y, w, h in detection:
  cv2.rectangle(image_color, (x, y), (x + w, y + h), (255,255,255))
cv2_imshow(image_color)
```

### Dlib

```python
import dlib
```

```python
image = cv2.imread('/content/drive/MyDrive/Deep Learning/work-space.png')

# Convert to gray
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#Change the scale
new_scale = tuple(round(i / 1.5) for i in reversed(image_gray.shape))
image_gray = cv2.resize(image_gray, new_scale)
image_color = cv2.resize(image, new_scale)
```

```python
detector_face_hog = dlib.get_frontal_face_detector()
detection = detector_face_hog(image_color,2)

for face in detection:
  left,top,right,bottom = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle(image_color, (left, top), (right, bottom), (255,255,255))
cv2_imshow(image_color)
```
